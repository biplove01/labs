## Data Loading
First, I will load the `rrn1.csv` dataset into a pandas DataFrame. This is the initial step to make the data accessible for analysis.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/rrn1.csv')

print("Dataset loaded successfully. Displaying the first 5 rows:")
display(df.head())
```

## Descriptive Statistics
Next, I will use descriptive statistics to summarize the data. This provides a quick overview of the central tendency, dispersion, and shape of the distribution of each attribute.

```python
print("Descriptive statistics for the dataset:")
display(df.describe())
```

## Extracting Mean and Median of Specific Attributes
Now, I will extract the mean and median for 'max temperature' and 'min atmos pressure' as requested. This helps in understanding the typical values for these key weather parameters.

```python
attributes_to_analyze = ['max.temp', 'min.atmos.pressure'] # Corrected column names

# Check if columns exist before processing
missing_cols = [col for col in attributes_to_analyze if col not in df.columns]
if missing_cols:
    print(f"Warning: The following columns were not found in the dataset and will be skipped: {missing_cols}")

print("\nExtracting Mean and Median for requested attributes from cleaned data:")
for attribute in attributes_to_analyze:
    if attribute in df.columns:
        mean_val = df[attribute].mean()
        median_val = df[attribute].median()
        print(f"\nAttribute: {attribute}")
        print(f"  Mean: {mean_val:.2f}")
        print(f"  Median: {median_val:.2f}")
    else:
        print(f"\nColumn '{attribute}' not found in the DataFrame. Please check the column name.")
```

## Handling Missing Values
Before proceeding with further analysis, it's crucial to handle any missing values in the dataset. First, I will identify the number of missing values per column. Then, I will fill numerical missing values with the mean of their respective columns and categorical missing values with the mode. This is a common imputation strategy to ensure data completeness without losing too much information.

```python
print("Number of missing values before handling:")
display(df.isnull().sum())

# Explicitly convert 'max.temp' to numeric, coercing 'NAN' strings to actual NaN values
# This needs to happen BEFORE identifying numerical_cols if 'max.temp' might be object type.
if 'max.temp' in df.columns:
    initial_dtype = df['max.temp'].dtype
    df['max.temp'] = pd.to_numeric(df['max.temp'], errors='coerce')
    if df['max.temp'].isnull().any() and initial_dtype == 'object':
        print("Converted 'max.temp' to numeric and recognized 'NAN' as missing values.")

# Identify numerical and categorical columns AFTER potential type conversion
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

print("\nMissing values AFTER potential type conversion (before imputation):")
display(df.isnull().sum())

# Impute missing values
# For numerical columns, fill with mean
for col in numerical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean())
        print(f"Filled missing values in numerical column '{col}' with its mean.")

# For categorical columns, fill with mode
for col in categorical_cols:
    if df[col].isnull().any():
        mode_val = df[col].mode()[0] # mode() can return multiple values, pick the first
        df[col] = df[col].fillna(mode_val)
        print(f"Filled missing values in categorical column '{col}' with its mode ({mode_val}).")

print("\nMissing values after handling:")
display(df.isnull().sum())
```

## Displaying the Cleaned Dataset
Finally, I will display the first 5 rows of the cleaned dataset. This shows the result of the data cleaning process, ensuring that missing values have been appropriately handled.

```python
print("Cleaned dataset (first 5 rows):")
display(df.head())
```
